# Introduction

## Installation and Running

The ASV simulation is split into multiple submodules.
When using the simulator for the first time or after an update from Git, you must first run the script `initialize.jl` to resolve local dependencies and precompile the packages.

```julia
include("initialize.jl")
```

## Module Overview

### EulerLagrangeX
A framework for modeling Euler-Lagrange systems with external inputs. It provides abstract types and interfaces for systems with generalized coordinates and velocities, including support for controlled systems. Calculates ordinary differential equations (ODEs) for Euler-Lagrange systems using automatic differentiation.

### CableModels
Implements rigid cable dynamics using the Euler-Lagrange framework. It models cables as a series of connected rigid segments with configurable mass and length properties. The module provides functions to compute cable positions, velocities, and accelerations at any point along the cable, as well as hydrodynamic forces and coordinate transformations. Cable dynamics are formulated using Lagrangian mechanics with support for external perturbations.

### MarineModels
Contains mathematical models and utilities specific to marine vehicles and environments. It includes a three degree-of-freedom ASV (Autonomous Surface Vehicle) model. The module also contains an abstract framework for implementing controllers for marine vehicles.

### PayloadModels
Contains a point-mass model of a towed payload.

### AdaptiveController
Implements the adaptive path-following towing controller for the ASV. It includes line-of-sight guidance and the adaptive linearizing controller (cf., `LinkedMasses`), together with a framework for making the control law applicable to the ASV.

### Simulator
The main simulation module that brings together all components. It implements the ODEs of the entire towed system, provides the default parameters used in the CEP paper, and contains utility functions for interpreting and plotting the simulation results.
